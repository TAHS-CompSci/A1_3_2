Defining Functions

```python
def add_tip(total, tip_percent):
    '''Return the total amount of a meal including tip'''
    tip=tip_percent*total
    return total + tip
```

```python
a = add_tip(20,0.15)
```

```python
a
```

```python
add_tip(30,0.15)
```

```python
add_tip(30)
```

```python
def hyp(leg1, leg2):
    '''Returnsthe hypotenuse'''
    number=(leg1*leg1+leg2*leg2)**0.5
    return number
```

```python
hyp(3,4)
```

```python
def mean(a,b,c):
    '''Returns the mean'''
    mean=(a+b+c)/3
    return mean
```

```python
mean(3,4,7)
```

```python
def perimeter(a,b):
    '''Returns the permimeter'''
    number=(a+a+b+b)
    return number
```

```python
perimeter(3,4)
```
